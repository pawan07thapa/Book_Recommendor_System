```python
#finding avg rreading of each book
average_rating=pd.DataFrame(ratings.groupby('ISBN')['bookRating'].mean())
average_rating['rating_count']=pd.DataFrame(ratings.groupby('ISBN')['bookRating'].count())
average_rating.sort_values('rating_count',ascending=False).head()
```

```python
#To ensure statistical significance, users with less than 200 ratings, and books with less than 100 ratings are excluded.

count1=ratings['userID'].value_counts() 
print(count1)

```

```python
ratings=ratings [ratings['userID'].isin( count1 [count1 >= 200] . index)]
ratings.head(7)
```

```python
counts=ratings['bookRating'].value_counts()
ratings=ratings[ratings['bookRating'].isin(counts[counts >= 100].index)]
```

```python
#RATING MATRIX

ratings_pivot=ratings.pivot(index='userID',columns='ISBN').bookRating  #isme error aa rhi hai
userId=ratings_pivot.index
ISBN=ratings_pivot.columns
print(ratings_pivot.shape)
ratings_pivot.head(5)
```

```python
bones_ratings=ratings_pivot['0316666343']
similar_to_bones=ratings_pivot.corrwith(bones_ratings)
corr_bones=pd.DataFrame(similar_to_bones,columns=['pearson R'])
corr_bones.dropna(inplace=True)
corr_summary=corr_bones.join(average_rating['rating_count'])
corr_summary[corr_summary['rating_count'] >=300 ].sort_values('pearson R',ascending=False).head(10)
```
